```python
# Your Answer

# spark.sql("select * from tmp_df_data   ").show(truncate=False)

dat_file='tickets.dat'
df1=spark.read.csv(dat_file,sep="|",inferSchema=True,header=True)

df1.columns
```

```python
df1.createOrReplaceTempView("tmp_df_data_tkt")
spark.sql("select priority,solved_by, COUNT(*) from tmp_df_data_tkt  group by  priority,solved_by  ").show(100)
# spark.sql("select case when priority= then ,solved_by, COUNT(*) from tmp_df_data_tkt  group by  priority,solved_by  ").show(100)


```

```python

sql_query="""
select
solved_by as team,
COUNT(case when priority='1'  THEN  ticket_id  ELSE NULL END)  AS  tickets_priority_1_count,
COUNT(case when priority='2'  THEN ticket_id  ELSE NULL END)  AS  tickets_priority_2_count,
COUNT(case when priority='3'  THEN ticket_id  ELSE NULL END)  AS  tickets_priority_3_count

FROM 
tmp_df_data_tkt
group BY solved_by

"""


spark.sql(sql_query).show(100)
```

```python

sql_df="""
select 
tkt.ticket_id,
tkt.alarm,
tkt.started_ts,
tkt.ended_ts,
tkt.solved_by,
tkt.src_system,
tkt.site_visit,
tkt.priority,
t.team_size
FROM tmp_df_data_tkt as tkt
LEFT JOIN (SELECT team_size , replace (team_id,'team0','team') as team_id FROM tmp_team_vw) as t ON (t.team_id=tkt.solved_by)
"""
spark.sql(sql_df).show(10)
```

```python
from pyspark.sql.functions import *
from  pyspark.sql.functions import lag
```

```python

sql_df="""


select 
tkt.solved_by
,t.team_size
,year(  started_ts) as year
,month(  started_ts) as month
,COUNT(*) as tickets_solved
FROM tmp_df_data_tkt as tkt
LEFT JOIN (SELECT team_size , replace (team_id,'team0','team') as team_id FROM tmp_team_vw) as t ON (t.team_id=tkt.solved_by)
where t.team_size=6
group by 
year(  started_ts)  ,
month(  started_ts) ,
tkt.solved_by,
t.team_size
order by solved_by,month
"""
spark.sql(sql_df).show(10)
```

```python

sql_df="""

select 
year,
month,
solved_by as team,
team_size,
tickets_solved as tickets_solved_in_the_current_month,
lag(tickets_solved) over (partition by solved_by,year order by month  ) tickets_solved_in_the_previous_year
FROM
(select 
tkt.solved_by
,t.team_size
,year(  started_ts) as year
,month(  started_ts) as month
,COUNT(*) as tickets_solved
FROM tmp_df_data_tkt as tkt
LEFT JOIN (SELECT team_size , replace (team_id,'team0','team') as team_id FROM tmp_team_vw) as t ON (t.team_id=tkt.solved_by)
group by 
year(  started_ts)  ,
month(  started_ts) ,
tkt.solved_by,
t.team_size
)a
"""
spark.sql(sql_df).show(10)
```
